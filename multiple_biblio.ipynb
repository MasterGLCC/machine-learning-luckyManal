```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# chargement du dataset depuis un fichier CSV
data = pd.read_csv("../data/data_nutrition.csv")

# définition des features
features = [
    'Year',
    'Dietary_Energy_Adequacy',
    'Prevalence_of_Undernourishment_in_percent',
    'Consumer_Price_Index_in_percent'
]

# définition de la variable cible c'est target (ce qu'on veut prédire)
target = 'Number_of_Undernourished_People_in_percent'

# séparation des données en X (features) et Y (target)
X = data[features]
Y = data[target]

# création du modèle de régression linéaire (sklearn) prêt à apprendre
model = LinearRegression()

# c'est dans cette étape ou se passe l'entraînement du modèle sur les données quand a préparé avant.
#fit calcule directement les coefficients optimaux du modèle en minimisant l'erreur globale du modèle
model.fit(X, Y)

# prédictions sur les mêmes données en appliquant les coefficients (poids) appris par la fonction fit
predictions = model.predict(X)

# affichage des coefficients appris par le modèle
print("Coefficients (poids) : ", model.coef_)
print("Biais : ", model.intercept_)

# prédiction sur un exemple qui n'est pas présent dans notre dataset
# en respectant la structure des données d'entraînement
exemple = pd.DataFrame([[2025, 6.3, 2.2, 120]], columns=features)

pred = model.predict(exemple)

print(f"Prédiction pour l'exemple qu'on a proposé: {pred[0]:.2f}% de personnes sous-alimentées")
```
