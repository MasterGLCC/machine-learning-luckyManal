```python
import pandas as pd

data = pd.read_csv("../data/data_nutrition.csv")
feature =  'Consumer_Price_Index_in_percent'
target = 'Number_of_Undernourished_People_in_percent'

#init_X = data[feature].tolist()  

# extraction de la feature:
# ici on choisi une seule colonne de notre dataset (CPI)
X = data[feature].tolist()  
# extraction de la variable cible(target)
# c'est ce qu'on veut prédire : le % de sous-alimentation
Y = data[target].tolist()


# transformation polynomiale (x, x²)
# permet de modéliser une courbe au lieu d'une droite
X_poly = [[x, x**2] for x in X]

# initialisation des paramètres du modèle
beta = [0.0, 0.0, 0.0] # beta0, beta1, beta2
learning_rate = 0.00000001
iterations = 5000
n = len(Y)

# boucle d'apprentissage (gradient descent)
for iteration in range(iterations):
    # liste ou on va stocker le gradient de chaque poids
    # au début on met des 0 car on va les calculer après
    gradients = [0.0, 0.0, 0.0]
    mse = 0.0
    
    # calcul des prédictions et des gradients
    for i in range(n):
        # pour chaque ligne :
        # on calcule la prédiction y_pred
        # par le modèle polynomial : y = beta0 + beta1 x + beta2 x²
        y_pred = beta[0] + beta[1]*X_poly[i][0] + beta[2]*X_poly[i][1]
        # on calcule l'erreur (y_pred - y réel)
        error = y_pred - Y[i]
        # accumulation de l'erreur quadratique
        mse += error**2
        # calcul des gradients pour chaque paramètre (beta) 
        gradients[0] += error
        gradients[1] += error * X_poly[i][0]
        gradients[2] += error * X_poly[i][1]
    # mise à jour des betas du modèle (on a range(3) car on a 3 beta : beta0, beta1 de x1 et beta2 de x1**2)
    for j in range(3):
        beta[j] -= (1/n) * learning_rate * gradients[j]
        
    # affichage de l'évolution de l'erreur
    if iteration % 500 == 0:
        print(f"Iteration {iteration} - MSE: {mse/(2*n):.4f}")

# prédictions finales avec le modèle appris
Y_pred = [beta[0] + beta[1]*x + beta[2]*x**2 for x in X]

# Observation :
# Lors de l'utilisation de cette feature, on observe une erreur de type overflow
# Cela est dû à l'échelle des données (valeurs trop grandes) et à l'absence de normalisation.
# j ai choisi learning_rate = 0.00000001 une valeur très petite de pas ce qui est pas optimale car le cout se 
#diminue lentement mais au moins on a pu voir des valeurs 
```
