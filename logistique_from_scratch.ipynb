```python
import pandas as pd
import math

# lire les données depuis le fichier
df = pd.read_csv("../data/heart.csv")  

# X = toutes les features sauf colonne target
X = df.drop("target", axis=1).values.tolist()

# extraction de la variable cible(target)
# c'est ce qu'on veut prédire : malade ou pas malade”
y = df["target"].tolist()

m = len(X)   # nombre de lignes(exemples)
n = len(X[0])  # nombre de features

# initialisation de la liste des poids (beta1, beta2, ...)
W = [0.0 for _ in range(n)]
# biais (beta0)
bias = 0.0
# learning rate
alpha = 0.0001
# nombre d'itérations
iterations = 1000

# j'ai fait fonction qui retourne la formule de sigmoid pour qu'elle soit facile à utiliser après
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# boucle d'apprentissage (gradient descent)
for iteration in range(iterations):
    # gradients
    dw = [0.0 for _ in range(n)]
    # gradient du bias
    dbias = 0.0
    # loss totale
    loss = 0.0
    
    # calcul des prédictions et des gradients
    for i in range(m):
        # modèle linéaire: z = beta0 + Σ betaj * xj
        # bias c'est le beta0 on commence par ça pour ajouter les autres élements
        z = bias
        for j in range(n):
            z += W[j] * X[i][j] 
        # transformé notre prédiction en probabilité
        y_prob = sigmoid(z)
        # appliquer la formule du log loss (on va diviser par m aprés :lors de l'affichage)
        loss += -(y[i] * math.log(y_prob + 1e-9) +
                  (1 - y[i]) * math.log(1 - y_prob + 1e-9))
        # erreur pour une seule ligne
        error = y_prob - y[i]
        # calcul de gradients 
        for j in range(n):
            dw[j] += error * X[i][j]

        dbias += error
        
    # mise à jour des poids
    for j in range(n):
        W[j] -= alpha * (dw[j] / m)
    # Update du bias
    bias -= alpha * (dbias / m)

    # affichage de l'évolution de l'erreur
    if iteration % 100 == 0:
        print("Iteration:", iteration, "Loss:", loss / m)


# j'ai crée une fonction simple de prediction juste pour tester mon modèle
def predict(x):
    z = bias
    for j in range(n):
        z += W[j] * x[j]
    prob = sigmoid(z)
    #ici j'ai fais la classification
    return 1 if prob >= 0.5 else 0
    
# test sur une ligne du dataset
sample = X[100]

print("Prediction:", predict(sample))
print("True value:", y[100])
#j'ai ajouté ce petit bloc de code juste pour voir à quelle point mon modèle est proche des valeurs réels
correct = 0
for i in range(m):
    # ici predict return target et y[i] contient la vraie valeur de target de cette ligne
    if predict(X[i]) == y[i]:
        correct += 1
# l'affichage de la moyenne
print("Accuracy:", correct / m)

#on remarque dans l'affichage que prediction et la valeur réelle sont identiques pour cette exemmple 
#la moyenne:(résultat ->Accuracy: 0.6673170731707317 ce qui est bon comme meme pour un code from scratch)          ce qui montre que notre modèle a bien appris 
```
