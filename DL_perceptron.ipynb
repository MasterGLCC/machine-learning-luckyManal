```python
import pandas as pd
import numpy as np

#Chargement des données
data = pd.read_csv("data/diabetes.csv")

# Séparation des données
# X = toutes les features 
# on enlève la dernière colonne car c’est la target
X = data.drop(columns=[data.columns[-1]]).values

# y = la target (ce qu’on veut prédire)
# on reshape pour avoir une colonne (format matrice)
y = data[data.columns[-1]].values.reshape(-1, 1)

# Initialisation du modèle
# n_samples = nombre de patients(lignes)
# n_features = nombre de caractéristiques par patient
n_samples, n_features = X.shape

# weights = importance de chaque feature
# on commence à zéro car le modèle ne sait rien au début(aléatoire)
weights = np.zeros((n_features, 1))
bias = 0

#vitesse d’apprentissage
learning_rate = 0.0001

# epochs = nombre de fois qu’on va parcourir tout le dataset
epochs = 1000

# epsilon = petite valeur pour éviter division par zéro
epsilon = 1e-15

# Fonction sigmoïde
# Ici on transforme un nombre Z en probabilité entre 0 et 1
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Entraînement du modèle
for epoch in range(epochs):

    # On calcule Z = XW + b
    # c’est le score brut du modèle
    Z = np.dot(X, weights) + bias

    # On applique la sigmoïde pour obtenir une probabilité
    # y_pred = estimation du modèle
    y_pred = sigmoid(Z)

    # Calcul de la loss
    # La log-loss mesure l’erreur entre y réel et y prédit
    # plus la loss est grande, plus le modèle est mauvais
    loss = -np.mean(
        y * np.log(y_pred + epsilon)
        + (1 - y) * np.log(1 - y_pred + epsilon)
    )

    # calcul des gradients
    # (différence entre prédiction et vérité)
    dLoss_dZ = y_pred - y

    # gradient des poids :
    # on regarde comment chaque feature influence l’erreur
    dLoss_dW = (1 / n_samples) * np.dot(X.T, dLoss_dZ)

    # gradient du biais :
    # moyenne de l’erreur globale
    dLoss_db = (1 / n_samples) * np.sum(dLoss_dZ)

    # Mise à jour des paramètres
    # on corrige les poids pour réduire l’erreur
    weights = weights - learning_rate * dLoss_dW
    bias = bias - learning_rate * dLoss_db

    # affichage tous les 100 epochs pour suivre l’apprentissage
    if epoch % 100 == 0:
        print(f"Epoch {epoch} | Current training loss = {loss:.6f}")

# Résultat final du modèle
print("\nPoids finaux :")
print(weights)

# Prédiction sur les données (évaluation du modèle)
# On refait le calcul Z = XW + b avec les poids finaux appris
# C’est le même calcul que pendant l’entraînement, mais ici on évalue le modèle
Z = np.dot(X, weights) + bias

# On applique la fonction sigmoïde pour obtenir des probabilités
# Chaque valeur représente la probabilité d'appartenir à la classe 1
y_pred = sigmoid(Z)

# On transforme les probabilités en classes (0 ou 1)
# règle : si probabilité >= 0.5 → classe 1 sinon classe 0
predictions = (y_pred >= 0.5).astype(int)

# On calcule l’accuracy
# c’est le pourcentage de bonnes prédictions par rapport aux vraies valeurs y
accuracy = np.mean(predictions == y)

# Affichage du résultat final du modèle
print("Accuracy :", accuracy)
```
